```python
import json
import random
import numpy as np
import pandas as pd
import plotly.offline as py
from datetime import datetime
import plotly.graph_objs as go
from pandas.io.json import json_normalize
```

```python
#parameter
with open('../dataset/labels_nmf.json',encoding = 'utf8') as file:
    data = json.load(file)
```

```python
#parameter
cluster_num = 30
cluster_output = data
chart_path = 'visualization_result/box-plot.html'
```

```python
#Load article data
with open('./dataset/dataset.json',encoding = 'utf8') as file:
    data = json.load(file)
df = json_normalize(data)
df['created_time'] = pd.to_datetime(df['created_time'])
df = df.set_index('created_time')
df = df.dropna(subset=['message'])
df.shape
```

```python
arr = [[] for i in range(cluster_num)]
for ctr, value in enumerate(cluster_output):
    ns = 1e-9
    dt = datetime.utcfromtimestamp(df.index.values[ctr].astype(int) * ns).date()
    arr[value].append(dt.replace(year = 2016))

df_cluster = pd.DataFrame(arr)
df_cluster = df_cluster.transpose()
df_cluster.describe()
```

```python
data = []

for col in df_cluster.columns:
    data.append(  go.Box( y=df_cluster[col], name=col, showlegend=False ) )

data.append( go.Scatter( x = df_cluster.columns, y = df_cluster.mean(), mode='lines', name='mean' ) )

url = py.plot(data, filename=chart_path)
```
